```python
import os
import numpy as np
import shutil
import random
import pandas as pd
```

# <font color='red'>**Kfold stratified**</font>
Train: 80%, Test: 20%
* Ade: are el 53% of the samples
* Hyp: are el 28% of the samples
* Ser: are el 19% of the samples

```python
#5 kfold splits (only test sets):
#k=1
kf1_ade = [9, 10, 14, 19, 21, 22, 33, 36]
kf1_hyp = [1, 3, 11, 13]

#k=2
kf2_ade = [8, 11, 15, 17, 23, 30, 37, 39]
kf2_hyp = [2, 4, 7, 15]

#k=3
kf3_ade = [1, 5, 7, 12, 16, 20, 24, 29]
kf3_hyp = [5, 8, 10, 12, 14]

#k=4
kf4_ade = [2, 4, 6, 13, 18, 25, 27, 31]
kf4_hyp = [6, 9, 16, 20]

#k=5
kf5_ade = [3, 26, 28, 32, 34, 35, 38, 40]
kf5_hyp = [17, 18, 19, 21]
```

```python
#for check if some elements are repited
adenomas = [kf1_ade, kf2_ade, kf3_ade, kf4_ade, kf5_ade]
hyperplas = [kf1_hyp, kf2_hyp, kf3_hyp, kf4_hyp, kf5_hyp]

print("for adenomas...")
for i in range(len(adenomas)):
    compare = adenomas.pop(i)
    for element in compare:
        for lista in adenomas:
            if element in lista:
                print("valor repetido: ", element, "in list: ", compare)
            else:
                None
    adenomas.append(compare)
    
print("for hyperplastics...")
for i in range(len(hyperplas)):
    compare = hyperplas.pop(i)
    for element in compare:
        for lista in hyperplas:
            if element in lista:
                print("valor repetido: ", element, "in list: ", compare)
            else:
                None
    hyperplas.append(compare)
```

```python
all_adenomas = range(1, 41, 1)
all_hyperplastics = range(1, 22, 1)

adenomas_test = [kf1_ade, kf2_ade, kf3_ade, kf4_ade, kf5_ade]
hyperplas_test = [kf1_hyp, kf2_hyp, kf3_hyp, kf4_hyp, kf5_hyp]
```

## <font color='red'>**Saving folds**</font>

```python
#amount of frames per class: adenoma: 19, hyperplastic: 38
source_path = '../../../../../data/polyp_original/'
save_path = '../../../data/binary/'
wl_path = source_path + 'WL/'
nbi_path = source_path + 'NBI/'
ade_can = 19
hyp_can = 38
random.seed(14)

for i, test in enumerate(adenomas_test):
    to_train = []
    print("==== for fold ", i+1, "=====\n")    
    for element in all_adenomas:
        if element not in test:
            to_train.append(element)
        else:
            None  
    
    #reading and saving test WL/NBI elements
    splits = [test, to_train]
    for j, split in enumerate(splits):
            
        for element in split:
            current_path_wl = wl_path + 'adenoma_WL/video_' + str(element)
            current_path_nbi = nbi_path + 'adenoma_NBI/video_' + str(element)
            #getting amount of frames by folder
            can_wl = len(os.listdir(current_path_wl))
            can_nbi = len(os.listdir(current_path_nbi))
            #getting the minimum amount of frames
            imgs_wl = random.sample(range(1,can_wl),ade_can)
            imgs_nbi = random.sample(range(1,can_nbi),ade_can)

            #reading and saving wl images 
            for img in imgs_wl:
                if img < 10:
                    file = '_img_000' + str(img) + '.png'
                elif img < 100:
                    file = '_img_00' + str(img) + '.png'
                elif img < 1000:
                    file = '_img_0' + str(img) + '.png'
                else:
                    file = '_img_' + str(img) + '.png'                    
                    
                pth_wl = wl_path+'adenoma_WL/video_'+str(element)+'/adenoma_WL_video_'+str(element)+file 
                
                if j == 0:
                    target_pth = save_path + 'fold' + str(i+1) + '/test_WL/'
                else:
                    target_pth = save_path + 'fold' + str(i+1) + '/train_WL/'
                
                #for make the directory if it does not exist
                if not os.path.exists(target_pth):
                    os.makedirs(target_pth)
                else:
                    None
                shutil.copy(pth_wl, target_pth)
                
            #reading and saving nbi images 
            for ima in imgs_nbi:
                if ima < 10:
                    file = '_img_000' + str(ima) + '.png'
                elif ima < 100:
                    file = '_img_00' + str(ima) + '.png'
                elif ima < 1000:
                    file = '_img_0' + str(ima) + '.png'
                else:
                    file = '_img_' + str(ima) + '.png' 
                    
                pth_nbi = nbi_path+'adenoma_NBI/video_'+str(element)+'/adenoma_NBI_video_'+str(element)+file 
                
                if j == 0:
                    target_pth = save_path + 'fold' + str(i+1) + '/test_NBI/'
                else:
                    target_pth = save_path + 'fold' + str(i+1) + '/train_NBI/'
                
                #for make the directory if it does not exist
                if not os.path.exists(target_pth):
                    os.makedirs(target_pth)
                else:
                    None
                    
                shutil.copy(pth_nbi, target_pth)   
                
print("finished!")
```
